# 使用第1类特征训练模型

````python


train['企业所得税与增值税之比'] = train['企业所得税']/train['增值税']
test['企业所得税与增值税之比'] = test['企业所得税']/test['增值税']
train['经营范围_2'] = train["经营范围"].map(lambda x : len(x))
test['经营范围_2'] = test["经营范围"].map(lambda x : len(x))
````

````python
# 'ID', '企业类型', '登记机关', '企业状态', '邮政编码', '注册资本', '核准日期', '行业代码', '经营期限自',
#        '成立日期', '行业门类', '企业类别', '管辖机关', '经营范围', '增值税', '企业所得税', '印花税', '教育费',
#        '城建税', 'Label'
feature_name_class1_uesed=['企业类型', '登记机关', '企业状态','行业门类', '企业类别', '管辖机关', '注册资本', 
        '增值税', '企业所得税', '印花税', '教育费', '城建税' , '企业所得税与增值税之比', '经营范围_2']
train[feature_name_class1_uesed].head(2)
````

````python
## lightgbm单模型
lgb_paras = {'objective': 'binary',
             'learning_rate': 0.1,
             'max_depth': 4 ,
             'feature_fraction': 0.8, 
             'bagging_fraction' : 0.8,
             'num_threads':-1}
train_x, test_x, train_y, test_y =train_test_split(train, train_label['Label'], test_size=0.2, random_state=2019)
lgb_model = lgb.train(lgb_paras,
                          train_set = lgb.Dataset(train_x[feature_name_class1_uesed], train_y),
                          valid_sets=[lgb.Dataset(test_x[feature_name_class1_uesed],test_y)],
                          num_boost_round=400,
                          early_stopping_rounds=50,
                          verbose_eval=50
                          )
y_val = lgb_model.predict(test_x[feature_name_class1_uesed], num_iteration=lgb_model.best_iteration)
print("roc_auc:", roc_auc_score(test_y, y_val))
````

````python
## catboost单模型
train_x, test_x, train_y, test_y =train_test_split(train, train_label['Label'], test_size=0.2, random_state=2019)
cat_model = CatBoostRegressor(iterations=100,
                              depth=5,
                              learning_rate=0.1, 
                              loss_function='RMSE', 
                              verbose=20, 
                              early_stopping_rounds=20)

cat_model.fit(train_x[feature_name_class1_uesed], train_y, 
             eval_set=(test_x[feature_name_class1_uesed], test_y))
y_val = cat_model.predict(test_x[feature_name_class1_uesed])
print("roc_auc:", roc_auc_score(test_y, y_val))
````

````python
## xgboost 单模型
train[feature_name_class1_uesed].info()
````

````python

object_col = ['企业类型', '登记机关','行业门类', '企业类别', '管辖机关']
for i in tqdm(object_col):
    lbl = LabelEncoder()
    train[i] = lbl.fit_transform(train[i].astype(str))
````

````python
train_x, test_x, train_y, test_y =train_test_split(train, train_label['Label'], test_size=0.2, random_state=2019)
xgb_model = XGBRegressor(learning_rate=0.1,
                              n_estimators=1000,
                              max_depth=5,
                              gamma=0,
                              verbosity=1,
                              subsample=0.8,
                            #   min_child_weight=1, 
                              objective='binary:logistic')
eval_set = [(train_x[feature_name_class1_uesed], train_y), (test_x[feature_name_class1_uesed], test_y)]
xgb_model.fit(train_x[feature_name_class1_uesed], train_y, 
        early_stopping_rounds = 20,
        eval_metric=["error", "logloss",'auc'], 
        eval_set=eval_set, verbose=20)
y_val = xgb_model.predict(test_x[feature_name_class1_uesed])
# print("roc_auc:", roc_auc_score(test_y, np.round(y_val)))
print("roc_auc:", roc_auc_score(test_y, y_val))
````

````python
# lightgbm 交叉验证模型

fi = []
cv_score = []
test_pred = np.zeros((test.shape[0],))
skf = StratifiedKFold(n_splits=5, random_state=2019, shuffle=True)
lgb_paras = {'objective': 'binary',
             'learning_rate': 0.1,
             'max_depth': 4 ,
             'feature_fraction': 0.8, 
             'bagging_fraction' : 0.8,
             'num_threads':-1}

for index, (train_index, test_index) in enumerate(skf.split(train, y)):
    print(index)
    train_x, test_x, train_y, test_y = train.iloc[train_index],train.iloc[test_index], y.iloc[train_index], y.iloc[test_index]
    lgb_model = lgb.train(lgb_paras,
                          train_set = lgb.Dataset(train_x[feature_name_uesd], train_y),
                          valid_sets=[lgb.Dataset(train_x[feature_name_uesd], train_y), lgb.Dataset(test_x[feature_name_uesd],test_y)],
                          num_boost_round=800 ,
                          verbose_eval=10,
#                           early_stopping_rounds=30,
#                           feval=roc_auc_score
                          )
    y_val = lgb_model.predict(test_x[feature_name_uesd], num_iteration=lgb_model.best_iteration)
    print( "test roc_auc",  roc_auc_score(test_y , y_val) )
    cv_score.append(roc_auc_score(test_y,y_val))
    print("cv_score:", cv_score[index])
    test_pred += lgb_model.predict(test[feature_name_uesd], num_iteration=lgb_model.best_iteration) / 5
    
submission = pd.read_csv('../data/submission.csv')
submission['Label'] = test_pred
submission.to_csv('../result/submission_lightgbm_feature_all_cv5.csv',index=False)
````

````python
# catboost 交叉验证模型

fi = []
cv_score = []
test_pred = np.zeros((test.shape[0],))
skf = StratifiedKFold(n_splits=5, random_state=2019, shuffle=True)



for index, (train_index, test_index) in enumerate(skf.split(train, y)):
    print(index)
    train_x, test_x, train_y, test_y = train.iloc[train_index],train.iloc[test_index], y.iloc[train_index], y.iloc[test_index]
    cat_model = CatBoostRegressor(iterations=300,
                              depth=5,
                              learning_rate=0.1, 
                              loss_function='RMSE', 
                              verbose=20, 
#                               early_stopping_rounds=20
                             )
    cat_model.fit(train_x[feature_name_uesd], train_y, 
             eval_set=(test_x[feature_name_uesd], test_y))
    y_val = cat_model.predict(test_x[feature_name_uesd])
    print("roc_auc:", roc_auc_score(test_y, y_val))
    cv_score.append(roc_auc_score(test_y,y_val))
    print("cv_score:", cv_score[index])
    test_pred += cat_model.predict(test[feature_name_uesd]) / 5
test_pred = [0.0 if r < 0  else r for r in test_pred]
submission = pd.read_csv('../data/submission.csv')
submission['Label'] = test_pred
submission.to_csv('../result/submission_catboost_feature_all_cv5.csv',index=False)
````

````python
# xgboost 交叉验证

object_col = ['企业类型', '登记机关','行业门类', '企业类别', '管辖机关']
for i in tqdm(object_col):
    lbl = LabelEncoder()
    train[i] = lbl.fit_transform(train[i].astype(str))

fi = []
cv_score = []
test_pred = np.zeros((test.shape[0],))
skf = StratifiedKFold(n_splits=5, random_state=2019, shuffle=True)


for index, (train_index, test_index) in enumerate(skf.split(train, y)):
    print(index)
    train_x, test_x, train_y, test_y = train.iloc[train_index],train.iloc[test_index], y.iloc[train_index], y.iloc[test_index]
    xgb_model = XGBRegressor(learning_rate=0.15,
                              n_estimators=400,
                              max_depth=5,
                              gamma=0,
                              verbosity=1,
                              subsample=0.8,
                            #   min_child_weight=1, 
                              objective='binary:logistic')
    eval_set = [(train_x[feature_name_uesd], train_y), (test_x[feature_name_uesd], test_y)]
    xgb_model.fit(train_x[feature_name_uesd], train_y, 
#         early_stopping_rounds = 20,
        eval_metric=["error", "logloss",'auc'], 
        eval_set=eval_set, verbose=20)
    
   
    y_val = xgb_model.predict(test_x[feature_name_uesd])
    print("roc_auc:", roc_auc_score(test_y, y_val))
    cv_score.append(roc_auc_score(test_y,y_val))
    print("cv_score:", cv_score[index])
    test_pred += xgb_model.predict(test[feature_name_uesd]) / 5

submission = pd.read_csv('../data/submission.csv')
submission['Label'] = test_pred
submission.to_csv('../result/submission_xgb_feature_all_cv5.csv',index=False)
````
